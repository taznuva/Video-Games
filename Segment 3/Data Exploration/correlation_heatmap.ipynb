```python
# Generate correlation map for dataset
corr = df.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

```python
cols = ['Platform', 'Main Genre', 'Developer', 'Critic_Score', 'User Score', 'Global Sales']
for col in cols:
    uniques = df[col].value_counts().keys()
    uniques_dict = {}
    ct = 0
    for i in uniques:
        uniques_dict[i] = ct
        ct += 1

    for k, v in uniques_dict.items():
        df.loc[df[col] == k, col] = v
```

```python
df1 = pd.get_dummies(df[['Platform','Main Genre','Critic_Score',
                         'User Score','Global Sales']])
df1 = df1.dropna().reset_index(drop=True)
df1 = df1.astype('float64')
```

```python
mask = np.zeros_like(df1.corr())
mask[np.triu_indices_from(mask)] = True
cmap = sns.diverging_palette(730, 300, sep=20, as_cmap=True, s=85, l=15, n=20) # note: 680, 350/470
with sns.axes_style("white"):
    fig, ax = plt.subplots(1,1, figsize=(15,10))
    ax = sns.heatmap(df1.corr(), mask=mask,vmax=0.2, square=True, annot=True, fmt=".3f", cmap="Blues")
```

```python
def score_group(score):
    if score >= 90:
        return '90-100'
    elif score >= 80:
        return '80-89'
    elif score >= 70:
        return '70-79'
    elif score >= 60:
        return '60-69'
    elif score >= 50:
        return '50-59'
    else:
        return '0-49'
```

```python
score_df = df.dropna(subset=['Critic_Score']).reset_index(drop=True)
score_df['score_group'] = score_df['Critic_Score'].apply(lambda x: score_group(x))
```

```python
franchise = pd.get_dummies(df['Franchise'])
```

```python
def in_top(x):
    if x in pack:
        return x
    else:
        pass
def width(x):
    if x == 'Platform':
        return 14.4
    elif x == 'Developer':
        return 13.2
    elif x == 'Franchise':
        return 13.4
    elif x == 'Main Genre':
        return 13.6

def height(x):
    if x == 'Main Genre':
        return 8
    else:
        return 9
```

```python
cols = ['Platform', 'Developer', 'Main Genre', 'Franchise']
for col in cols:
    pack = []
    top = score_df[['Game Name', col]].groupby([col]).count().sort_values('Game Name', ascending=False).reset_index()[:15]
    for x in top[col]:
        pack.append(x)
    score_df[col] = score_df[col].apply(lambda x: in_top(x))
    score_df_platform = score_df[[col, 'score_group', 'Global Sales']].groupby([col, 'score_group']).median().reset_index().pivot(col, "score_group", "Global Sales")
    plt.figure(figsize=(width(col), height(col)))
    sns.heatmap(score_df_platform, cmap="Blues", annot=True, fmt=".2g", linewidths=.5).set_title((' \n'+col+' vs. Critic Score (by median sales) \n'), fontsize=18)
    plt.ylabel('', fontsize=14)
    plt.xlabel('score group \n', fontsize=14)
    pack = []
```
