```python
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter
```

```python
columns = [
    "Name", "platform", "Year_of_Release", "Genre",
    "Publisher", "NA_Salesr", "EU_Sales","JP_Sales",
    "Other_Sales","Global_Sales","Critic_Score","Critic_Count","User_Score","User_Count","Developer","Rating"
]

target = ["Global_Sales"]
```

```python
# Load the data
file_path = Path('/Users/Sheem/OneDrive/Desktop/Analysis Projects/Module 17/vgsales_clean.csv')
df = pd.read_csv(file_path, skiprows=0)



# Drop the null rows
df = df.dropna()

df.head(10)
```

```python
# Create our features
X = pd.get_dummies(df.drop('Global_Sales', axis=1))

# Create our target
y = df['Developer']
```

```python
X.describe()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=1)
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)

Counter(y_resampled)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=1, solver='lbfgs' )
model.fit(X_resampled, y_resampled)
```

```python
y_pred = model.predict(X_test)

results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
from sklearn.metrics import balanced_accuracy_score

balanced_accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = model.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, y_pred))
```
